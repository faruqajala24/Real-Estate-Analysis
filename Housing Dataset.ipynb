### Analyzing variables individually
First, let's define the names of the variables that we are going to use in this analysis. We have a list of the numerical variables and a list of the categorical variables. Then, we will redefine our housing DataFrame with a DataFrame that contains only the variables that we just defined.

```python
numerical_vars = ['TotRms AbvGrd',"Lot Area","Overall Qual","Overall Cond","Year Built","1st Flr SF","2nd Flr SF","Bedroom AbvGr"]
categorical_vars =["MS SubClass","MS Zoning","Street","Lot Shape","Neighborhood","Central Air","Sale Condition","Mo Sold","Yr Sold",'Sale Type']
```

```python
df_1 = df[numerical_vars+categorical_vars]
```

```python
df_1.shape
```

# Exploratory Data Analysis (EDA)

```python
# Generate summary statistics for numerical variables in the dataset.
df_1[numerical_vars].describe()
```

```python
# Create histograms to visualize the distribution of numerical variables.
df_1[numerical_vars].hist(edgecolor='black', bins=15, figsize=(14,5), layout=(2,4));
```

### The following is what we observe for each histogram:

The distribution of sizes in the first floor, as you can see, is skewed to the right. So what it tells you is that you have very few big houses, but most of the houses have these values around 1,000 or 1,200 square feet.
You see a big bar around 0 in the 2ndFlrSF variable. Those are the variables that don't have a second floor.
Most houses have three bedrooms.
The lot area is highly skewed; there are few houses with a large amount of area.
The ratings for conditions and quality tend to be around 5; few houses have very high or low ratings.
The YearBuilt variable is actually not useful in its present form. However, we can use it to construct a variable that actually makes sense; the age of the house at the time of the sale

```python
# Calculate the age of each property by subtracting the "Year Built" from the "Yr Sold" values.
df_1["Age"] = df_1["Yr Sold"] - df_1["Year Built"]
```

```python
# Check if "Year Built" is in the list before removing it
if "Year Built" in numerical_vars:
    numerical_vars.remove("Year Built")
# Append "Age" to the list
numerical_vars.append("Age")
```

```python
# Create histograms for a set of numerical variables in the dataset.
df_1[numerical_vars].hist(edgecolor='black', bins=15, figsize=(14, 5), layout=(2, 4));
```

### Here we see that the Age variable with a big bar here at 0 , so this means that many houses were sold; almost 400 houses were sold when they were new.

```python
# Create a bar plot to visualize the distribution of sale conditions in the dataset.
df_1["Sale Condition"].value_counts().plot(kind='bar', title="Sale Condition");
```

### When we execute the line, what we get is a count of the different categories within this Sale Condition variable. So, most of the houses were sold under the Normal condition, almost 1,200 of them, and very few houses were sold with the other conditions mentioned.

```python
import matplotlib.pyplot as plt
# Create a grid of subplots
fig, ax = plt.subplots(2,4, figsize=(14,6))

# Plot the value counts for each categorical variable
for var, subplot in zip(categorical_vars, ax.flatten()):
    df_1[var].value_counts().plot(kind='bar', ax=subplot,title=var)
    
# Add spacing between subplots for better readability    
fig.tight_layout()
```

### CONCLUSION


In this analysis of the housing dataset, we explored various aspects of residential properties, excluding the sale price, to gain insights into their characteristics. The dataset provided valuable information about different features, allowing us to understand the housing market and its dynamics.

Key Findings:

Zoning Classification: The dataset included various zoning classifications, with the majority of properties falling into the Residential Low Density (RL) category. Residential High Density (RH) and Residential Medium Density (RM) were also common.

Street Access: There were two types of road access to the properties, including Gravel (Grvl) and Paved (Pave). The majority of properties had paved access.

Dwelling Type: The dataset featured diverse types of dwellings, indicating a mix of housing styles within the area.

Condition: The overall condition of properties was assessed, providing insights into their maintenance and upkeep.

Year Built: The dataset contained information about the construction year of the properties, which can be valuable for understanding the age of the structures.

Amenities: The presence of additional amenities, such as pools, fences, and miscellaneous features, was also documented.

Age: We calculated the age of properties by subtracting the year built from the year sold, helping us understand the age distribution of the properties.

Data Exploration: Exploratory data analysis revealed patterns and trends within the dataset, shedding light on various property characteristics.

Implications:

The findings from this analysis can provide valuable insights for real estate professionals, property developers, and individuals interested in the local housing market. Understanding the distribution of zoning classifications, street access types, and dwelling types can inform decisions related to property development and investments.

Limitations:

It's essential to note that this analysis excluded sale price, a critical factor in the housing market. Sale price often plays a central role in real estate decisions, and its absence limited our ability to draw definitive conclusions about property values.

In summary, this analysis has provided a comprehensive overview of the housing dataset, offering valuable insights into the types and characteristics of properties within the area. Future studies may benefit from incorporating sale price data to gain a more complete understanding of the local housing market.
